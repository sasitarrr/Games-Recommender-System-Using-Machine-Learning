```python
rating_with_game = ratings.merge(name_img, on='game_name')
```

```python
num_rating = rating_with_game.groupby('game_name')['rating'].count().reset_index()
```

```python
num_rating.head()
```

```python
num_rating.rename(columns={
    'rating':'num_of_rating'},inplace=True)
```

```python
num_rating.head()
```

```python
rating_with_game.head()
```

```python
final_rating = rating_with_game.merge(num_rating, on='game_name')
```

```python
final_rating.head()
```

```python
final_rating.shape
```

```python
final_rating = final_rating[final_rating['num_of_rating']>=12]
```

```python
final_rating.head()
```

```python
final_rating.sample(5)
```

```python
final_rating.shape
```

```python
final_rating.drop_duplicates(['user_id','game_name'],inplace=True)
```

```python
final_rating.shape
```

```python
final_rating
```

```python
#7:Converting to Pivot Table
```

```python
game_pivot = ratings.pivot_table(columns='user_id', index='game_name', values='rating')
```

```python
game_pivot
```

```python
game_pivot.shape
```

```python
game_pivot.fillna(0,inplace=True)
```

```python
game_pivot
```

```python
from scipy.sparse import csr_matrix
```

```python
game_sparse = csr_matrix(game_pivot)
```

```python
game_sparse
```

```python
#8:Model Building
```

```python
from sklearn.neighbors import NearestNeighbors
model = NearestNeighbors(algorithm='brute')
```

```python
model.fit(game_sparse)
```

```python
distance, suggestion = model.kneighbors(game_pivot.iloc[12,:].values.reshape(1,-1), n_neighbors=4)
```

```python
distance
```

```python
suggestion
```

```python
for i in range(len(suggestion)):
    print(game_pivot.index[suggestion[i]])
```

```python
import pickle
pickle.dump(model, open('game_pickle/model.pkl', 'wb'))
pickle.dump(game_name, open('game_pickle/game_name.pkl', 'wb'))
pickle.dump(final_rating, open('game_pickle/final_rating.pkl', 'wb'))
pickle.dump(game_pivot, open('game_pickle/game_pivot.pkl', 'wb'))
```

```python
def recommend_game(game_name):
    game_id = np.where(game_pivot.index == game_name)[0][0]
    distance, suggestion = model.kneighbors(game_pivot.iloc[game_id,:].values.reshape(1,-1), n_neighbors=4)
    
    for i in range(len(suggestion)):
        games = game_pivot.index[suggestion[i]]
        for j in games:
            print(j)
```
